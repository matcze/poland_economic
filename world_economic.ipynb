```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv("gpd_per_capita.csv")
```

```python
dataset.head()
```

Poland’s economic growth over the years

```python
poland_data = dataset[dataset["Country Name"] == "Poland"]
years = ["2000", "2001", "2002", "2003", "2004", "2005", "2006", "2007", "2008","2009", "2010", "2011", "2012", "2013", "2014", "2015", "2016", "2017", "2018", "2019", '2020', "2021", "2022", "2023", "2024"]
```

```python
poland_centry = []

for year in years:
    poland_centry += poland_data[year].values.tolist()
```

```python
plt.figure(figsize=(11, 6))
plt.plot(years, poland_centry,  label="Poland", marker="s")

plt.xlabel("Year", fontsize=14, labelpad=10)
plt.ylabel("GDP", fontsize=14, labelpad=14)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show() 
```

A comparison of GDP per capita among Poland, Germany, and France.

```python
germany_data = dataset[dataset["Country Name"] == "Germany"]
france_data = dataset[dataset["Country Name"] == "France"]
```

```python
germany_centry = []
france_centry = []

for year in years:
    germany_centry += germany_data[year].values.tolist()
    france_centry += france_data[year].values.tolist()
```

```python
plt.figure(figsize=(11, 6))
plt.plot(years, germany_centry, label="Germany", marker="o")
plt.plot(years, poland_centry,  label="Poland", marker="s")
plt.plot(years, france_centry,  label="France", marker="D")

plt.xlabel("Year", fontsize=14, labelpad=10)
plt.ylabel("GDP", fontsize=14, labelpad=14)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show() 

```

Difference between Poland and Germany in GDP per capita in the 21st century.

```python
poland_centry_arr = np.array(poland_centry)
germany_centry_arr = np.array(germany_centry)

difference = germany_centry_arr - poland_centry_arr
print(difference)

plt.figure(figsize=(10, 6))
plt.bar(years, difference)
plt.title('Difference (Germany - Poland)')
plt.xlabel("Year", fontsize=14, labelpad=10)
plt.ylabel("GDP", fontsize=14, labelpad=14)
plt.grid(True, axis="y", linestyle='--', alpha=0.6)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

How did GDP per capita change in Poland during the governance of PO (2007–2015) and PiS (2015–2023)?

```python
fig, ax = plt.subplots(1, 2)
fig.set_size_inches((14, 5))
ax[0].plot(years[7:16], poland_centry[7:16], marker="o")
ax[0].set_xlabel("Year", fontsize=14, labelpad=10)
ax[0].set_ylabel("GDP", fontsize=14, labelpad=14)
ax[0].set_title('Poland governed by PO')
ax[0].grid(True, axis="y", linestyle='--', alpha=0.6)
plt.setp(ax[0].get_xticklabels(), rotation=45)



ax[1].plot(years[15:24], poland_centry[15:24], marker="o")
ax[1].set_title('Poland governed by PIS')
ax[1].set_xlabel("Year", fontsize=14, labelpad=10)
ax[1].set_ylabel("GDP", fontsize=14, labelpad=14)
ax[1].grid(True, axis="y", linestyle='--', alpha=0.6)
plt.setp(ax[1].get_xticklabels(), rotation=45)


plt.tight_layout()
plt.show()

```

GPD in Euro Area depends on country in 2024 year!

```python
eu_countries = ["Austria", "Belgium", "Bulgaria", "Croatia", "Cyprus", "Czechia", "Denmark", "Estonia", "Finland", "France", "Germany", "Greece", "Hungary", "Ireland", "Italy", "Latvia", "Lithuania", "Luxembourg", "Malta", "Netherlands", "Poland", "Portugal", "Romania", "Slovenia", "Spain", "Sweden"]
```

```python
dataset_2024 = dataset[["2024", "Country Name"]]
euro_gpd = []
for country in eu_countries:
    euro_gpd += dataset_2024[dataset["Country Name"] == country].values.tolist()

euro_gpd = np.array(euro_gpd)

sorted_indices = np.argsort(euro_gpd[:, 0].astype(float))[::-1]
euro_gpd_sorted = euro_gpd[sorted_indices]

countries = euro_gpd_sorted[:, 1]
gpd_values = euro_gpd_sorted[:, 0].astype(float)
highlight = "Poland"

colors = ['teal' if country != highlight else 'yellowgreen' for country in countries]

plt.figure(figsize=(12, 6))
bars = plt.bar(countries, gpd_values, color=colors)

for bar, country in zip(bars, countries):
    if country == highlight:
        bar.set_linewidth(1.5)
        bar.set_edgecolor('black')

plt.xticks(rotation=90)
plt.title("GDP per Capita in Euro Area (2024)")
plt.ylabel("GDP per Capita")
plt.tight_layout()
plt.show()
```

Comparison of USA and China’s Economic Growth

```python
usa_data = dataset[dataset["Country Name"] == "United States"]
china_data = dataset[dataset["Country Name"] == "China"]

usa_centry = []
china_centry = []

for year in years:
    usa_centry += usa_data[year].values.tolist()
    china_centry += china_data[year].values.tolist()
```

```python
plt.figure(figsize=(11, 6))
sns.lineplot(
    x=years, y=usa_centry,
    color="#1f77b4", label="United States of America",
    marker="o", linewidth=2.5, markersize=8
)
sns.lineplot(
    x=years, y=china_centry,
    color="#ff7f0e", label="China",
    marker="s", linewidth=2.5, markersize=8
)


sns.set_style("whitegrid")


plt.xlabel("Year", fontsize=14, labelpad=10)
plt.ylabel("GDP", fontsize=14, labelpad=14)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(fontsize=12)
plt.xticks(rotation=45, fontsize=12)
plt.yticks(fontsize=12) 
plt.tight_layout()
plt.show() 
```

Top 10 Richest Countries in the world

```python
dataset_richest = dataset_2024.sort_values("2024", ascending=False).head(15)
```

```python
sns.set_theme(style="whitegrid")
colors = sns.color_palette("Purples_r", len(dataset_richest) + 5)
plt.figure(figsize=(11, 6))

bars = plt.barh(dataset_richest["Country Name"], dataset_richest["2024"], color=colors, edgecolor="black")

for bar in bars:
    width = bar.get_width()
    plt.text(width * 1.02, bar.get_y() + bar.get_height()/2,
             f"{width:.2f}", va='center', fontsize=10)

plt.gca().invert_yaxis()
plt.title("Top 10 Richest Countries in the World", fontsize=16, weight='bold')

plt.xlabel("GPD per Capita", fontsize=12)
plt.ylabel("Country", fontsize=12)
plt.xticks(fontsize=12)  
plt.yticks(fontsize=12) 
plt.xlim(0, 160000)
plt.tight_layout()
plt.show()
```
